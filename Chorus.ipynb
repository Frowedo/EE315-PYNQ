# Chorus Effect

 The cell below creates a sin audio signal the progressivly compresses as a function of t^2

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data = np.sin(2*np.pi*220*t**2)
Audio(data,rate=framerate)
```

The cell below creates multiple sin audio signals all of varying amplitudes and frequencies. the signals are then summed together to create an overall chorus audio signal

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data = np.sin(2*np.pi*250*t**2)

data1 = 4*np.sin(2*np.pi*500*t**2)

data2 = 8*np.sin(2*np.pi*120*t**2)

data3 = -1*np.sin(2*np.pi*170*t**2)

data4 = -7*np.sin(2*np.pi*230*t**2)

SumData = data+data1+data2+data3+data4

Audio(SumData,rate=framerate)
```

The cells below plot the various seperate sin signals in order to observe their various amplitudes and frequencies

```python
import matplotlib.pylab as plt
plt.plot(t, data)
plt.title("Data")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -1, 1])
plt.show()
```

```python
import matplotlib.pylab as plt
plt.plot(t, data1)
plt.title("Data 1")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -4, 4])
plt.show()
```

```python
import matplotlib.pylab as plt
plt.plot(t, data2)
plt.title("Data 2")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -8, 8])
plt.show()
```

```python
import matplotlib.pylab as plt
plt.plot(t, data3)
plt.title("Data 3")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -1, 1])
plt.show()
```

```python
import matplotlib.pylab as plt
plt.plot(t, data4)
plt.title("Data 4")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -7, 7])
plt.show()
```

The cell below shows the waveform of the summed audio signal demonstrating the implemented chorus effect

```python
import matplotlib.pylab as plt
plt.plot(t, SumData)
plt.title("Sum Data")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.2, -20, 20])
plt.show()
```
