# Interference Effect

 The cell below creates a sin audio signal the progressivly compresses as a function of t^2

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data = np.sin(2*np.pi*250*t**2)
Audio(data,rate=framerate)
```

The cell below plots the sin audio signal of the original audio signal

```python
import matplotlib.pylab as plt
plt.plot(t, data)
plt.title("Original Data")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.4, -2, 2])
plt.show()
```

The cell below implements complete destructive interference where a signal is summed with its inverse thus the two signals cancel out to produce a 0 amplitude signal

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data = np.sin(2*np.pi*250*t**2)
data1 = -1*np.sin(2*np.pi*250*t**2)
dataSum = data+data1
Audio(dataSum,rate=framerate)
```

The cell below plots the sin audio signal of the complete destructive audio signal

```python
import matplotlib.pylab as plt
plt.plot(t, dataSum)
plt.title("Complete Destructive Interference")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.4, -2, 2])
plt.show()
```

The cell below implements constructive and destructive interference where a signal is summed with a signal of a different ampolitude and frequency, thus constructive interference occurs at some point in the signal and likewise with destructive interference

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data2 = np.sin(2*np.pi*250*t**2)
data3 = -0.9*np.sin(2*np.pi*230*t**2)
dataSum1 = data2+data3
Audio(dataSum1,rate=framerate)
```

The cell below plots the sin audio signal of the constructive and destructive audio signal

```python
import matplotlib.pylab as plt
plt.plot(t, dataSum1)
plt.title("Partial Destructive Interference")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.4, -2, 2])
plt.show()
```

The cell below implements constructive interference where a signal is summed with itself in order to increase the amplitude of the signal unifromly

```python
import numpy as np
from IPython.display import Audio
framerate = 44100
t = np.linspace(0,5,framerate*5)
data4 = np.sin(2*np.pi*250*t**2)
data5 = np.sin(2*np.pi*250*t**2)
dataSum2 = data4+data5
Audio(dataSum2,rate=framerate)
```

The cell below plots the sin audio signal of the constructive audio signal

```python
import matplotlib.pylab as plt
plt.plot(t, dataSum2)
plt.title("Constructive Interference")
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.axis([0, 0.4, -3, 3])
plt.show()
```
